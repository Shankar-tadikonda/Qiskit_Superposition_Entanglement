```python
# ===========================
# Install Required Libraries
# ===========================
!pip install qiskit qiskit-aer pylatexenc --quiet

# ===========================
# Import Libraries
# ===========================
%matplotlib inline
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
from IPython.display import display

# ===========================
# Task 1: Superposition
# ===========================
print("🔹 Task 1: Superposition")

# Create single-qubit circuit
qc_super = QuantumCircuit(1)
qc_super.h(0)                     # Apply Hadamard gate
qc_super.measure_all()             # Measure

# Statevector before measurement
state_super = Statevector.from_instruction(qc_super.remove_final_measurements(inplace=False))

# Display Bloch sphere
display(plot_bloch_multivector(state_super, title="Bloch Sphere – Superposition State"))

# Simulate
sim = AerSimulator()
result_super = sim.run(qc_super, shots=1024).result()
counts_super = result_super.get_counts()

# Display histogram
display(plot_histogram(counts_super, title="Superposition Measurement Results"))
print("Measurement counts:", counts_super)
print("\nExpected: roughly equal probability for 0 and 1\n")
```

```python
# ===========================
# Task 2: Entanglement
# ===========================
print("🔹 Task 2: Entanglement")

# Create 2-qubit circuit
qc_ent = QuantumCircuit(2)
qc_ent.h(0)        # Hadamard on qubit 0
qc_ent.cx(0, 1)    # CNOT to entangle 0 → 1
qc_ent.measure_all()

# Statevector before measurement
state_ent = Statevector.from_instruction(qc_ent.remove_final_measurements(inplace=False))

# Display Bloch spheres
display(plot_bloch_multivector(state_ent, title="Bloch Spheres – Entangled State"))

# Simulate
result_ent = sim.run(qc_ent, shots=1024).result()
counts_ent = result_ent.get_counts()

# Display histogram
display(plot_histogram(counts_ent, title="Entanglement Measurement Results"))
print("Measurement counts:", counts_ent)
print("\nExpected: correlated outcomes (00 or 11)\n")
```
